## Sprint 1: Listas

### 1. Propiedades básicas

```
mi_lista = [1, 2, 3, 4, 5]
```

### 2. Índices y slices

```
print(mi_lista[0])  # Salida: 1
```

```
print(mi_lista[1:4])  # Salida: [2, 3, 4]
```

### 3. Modificación de listas

```
mi_lista.append(6)  # mi_lista ahora es [1, 2, 3, 4, 5, 6]
```

```
mi_lista[1] = 20  # mi_lista ahora es [1, 20, 3, 4, 5]
```

```
mi_lista.remove(20)  # mi_lista ahora es [1, 3, 4, 5]
```

```
del mi_lista[2]  # mi_lista ahora es [1, 3, 5]
```

### 4. Ordenar listas

```
mi_lista.sort()  # Ordena la lista de forma ascendente
```

```
mi_lista.sort(reverse=True)  # Ordena la lista de forma descendente
```

### 5. Listas anidadas

```
lista_anidada = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
```

```
print(lista_anidada[1][2])  # Salida: 6
```

### 6. Procesamiento de strings

```
cadena = "Hola mundo"
lista_palabras = cadena.split()  # Salida: ['Hola', 'mundo']
```

```
nueva_cadena = " ".join(lista_palabras)  # Salida: 'Hola mundo'
```

```
cadena = "Hola Mundo"
print(cadena.lower())  # Salida: 'hola mundo'
```

```
print(cadena.upper())  # Salida: 'HOLA MUNDO'
```

```
nueva_cadena = cadena.replace("Mundo", "Python")  # Salida: 'Hola Python'
```

## Sprint 2: Bucles y Sentencias Condicionales

### 1. Bucles

#### a. Operadores de asignación aumentada y funciones integradas

```
suma = 0
for i in range(1, 11):
    suma += i  # Equivalente a suma = suma + i
print(suma)  # Salida: 55
```

#### b. Bucles `for`

```
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    if numero % 2 == 0:
        print(numero)  # Salida: 2, 4
```

#### c. Bucles sobre listas anidadas

```
lista_anidada = [[1, 2], [3, 4], [5, 6]]
for sublista in lista_anidada:
    for elemento in sublista:
        print(elemento)
```

#### d. Bucles `while`

```
contador = 1
while contador <= 5:
    print(contador)
    contador += 1
```

### 2. Sentencias condicionales

#### a. Expresiones lógicas y operadores de comparación

```
numero = 15
print(10 < numero < 20)  # Salida: True
```

#### b. Sentencias `if`, `else`, y `elif`

```
numero = -5
if numero > 0:
    print("Positivo")
else:
    print("Negativo")  # Salida: Negativo
```

```
nota = 85
if nota < 60:
    print("Reprobado")
elif 60 <= nota < 80:
    print("Aprobado")
else:
    print("Sobresaliente")  # Salida: Sobresaliente
```

#### c. Modificación de tablas y filtros

```
estudiantes = [
    {"nombre": "Ana", "calificacion": 85},
    {"nombre": "Juan", "calificacion": 78},
    {"nombre": "Luis", "calificacion": 90}
]

# Filtrar estudiantes con calificación mayor a 80
estudiantes_aprobados = [e for e in estudiantes if e["calificacion"] > 80]
print(estudiantes_aprobados)  # Salida: [{'nombre': 'Ana', 'calificacion': 85}, {'nombre': 'Luis', 'calificacion': 90}]
```

#### d. Operadores lógicos y filtros

```
numero = 45
if (numero < 30 or numero > 50) or (numero % 5 == 0):
    print("Condición cumplida")  # Salida: Condición cumplida
```
