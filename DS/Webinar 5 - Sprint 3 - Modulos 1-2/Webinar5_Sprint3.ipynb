```python
# 17. Eliminar filas con valores nulos
dropped_na_df = large_df_expanded.dropna(axis=0)
dropped_na_df.head()
```

#### Ejercicio 5: Trabajando con datos duplicados/ausentes

Los datasets nunca vienen limpios. Qué pasa cuando tenemos valores duplicados o vacios?

<div style="text-align: center;">
    <img src="dv.jpg" alt="House pricing" style="width:50%;"/>
</div>

```python
# Carguemos el dataset
import pandas as pd
currency_df = pd.read_csv('divisa.csv')
```

```python
currency_df.head()
```

```python
# 1. Resumen del DataFrame
currency_df.info()
```

```python
# 2. Contar valores nulos en cada columna
currency_df.isna().sum()
```

```python
# 2. Contar valores nulos en cada columna
currency_df.isnull().sum()
```

```python
# 3. Contar valores por categoría, incluyendo NaN
currency_df["Source"].value_counts(dropna=False)
```

```python
# 4. Ordenar por índice
currency_df.sort_index()
```

```python
# 5. Filtrar datos con condiciones
filtered_data = currency_df[
    (~currency_df["Currency Pair"].isna()) & (currency_df["Source"] == "Trading Platform A")
]
```

```python
filtered_data.head(20)
```

```python
# 6. Reemplazar valores vacíos
currency_df["Source"] = currency_df["Source"].replace("", "Unknown")
print(currency_df["Source"].unique())
```

```python
# 7. Rellenar valores nulos en `Volume` con la media
currency_df["Volume"] = currency_df["Volume"].fillna(currency_df["Volume"].mean())
print(currency_df["Volume"].head())
```

```python
# 8. Eliminar duplicados y restablecer índice
deduplicated_df = currency_df.drop_duplicates().reset_index(drop=True)
print(deduplicated_df.head())
```
