```python
import pandas as pd
import numpy as np
from scipy import stats as st
import matplotlib.pyplot as plt

# Generar datos ficticios para simular el dataset
np.random.seed(42)

# Crear 500 usuarios con planes aleatorios y ciudades
user_ids = np.arange(1, 501)
plans = np.random.choice(["Surf", "Ultimate"], size=500, p=[0.6, 0.4])
cities = np.random.choice(["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"], size=500)

# Fechas de registro y churn aleatorias
reg_dates = pd.date_range(start="2018-01-01", end="2018-06-30", periods=500)
churn_dates = np.random.choice([None, pd.NaT], size=500, p=[0.7, 0.3])

# Crear DataFrame de usuarios
users = pd.DataFrame({
    "user_id": user_ids,
    "plan": plans,
    "city": cities,
    "reg_date": reg_dates,
    "churn_date": churn_dates
})

# Generar llamadas
call_ids = np.arange(1, 5001)
call_dates = pd.date_range(start="2018-01-01", end="2018-12-31", periods=5000)
durations = np.random.randint(1, 20, size=5000)  # Duración entre 1 y 20 min
user_id_calls = np.random.choice(user_ids, size=5000)

calls = pd.DataFrame({
    "id": call_ids,
    "call_date": call_dates,
    "duration": durations,
    "user_id": user_id_calls
})

# Generar SMS
message_ids = np.arange(1, 2001)
message_dates = pd.date_range(start="2018-01-01", end="2018-12-31", periods=2000)
user_id_messages = np.random.choice(user_ids, size=2000)

messages = pd.DataFrame({
    "id": message_ids,
    "message_date": message_dates,
    "user_id": user_id_messages
})

# Generar uso de internet
internet_ids = np.arange(1, 3001)
session_dates = pd.date_range(start="2018-01-01", end="2018-12-31", periods=3000)
mb_used = np.random.randint(100, 3000, size=3000)  # Datos entre 100MB y 3GB
user_id_internet = np.random.choice(user_ids, size=3000)

internet = pd.DataFrame({
    "id": internet_ids,
    "session_date": session_dates,
    "mb_used": mb_used,
    "user_id": user_id_internet
})

# Crear DataFrame de tarifas
plans_data = pd.DataFrame({
    "plan_name": ["Surf", "Ultimate"],
    "usd_monthly_fee": [20, 70],
    "minutes_included": [500, 3000],
    "messages_included": [50, 1000],
    "mb_per_month_included": [15360, 30720],  # 15GB y 30GB en MB
    "usd_per_minute": [0.03, 0.01],
    "usd_per_message": [0.03, 0.01],
    "usd_per_gb": [10, 7]
})

# Guardar datasets ficticios como CSV
users.to_csv("megaline_users.csv", index=False)
calls.to_csv("megaline_calls.csv", index=False)
messages.to_csv("megaline_messages.csv", index=False)
internet.to_csv("megaline_internet.csv", index=False)
plans_data.to_csv("megaline_plans.csv", index=False)


print("Los archivos ficticios han sido generados y guardados. Ahora procederé a crear el notebook Jupyter.")
```
