```python
def get_products_by_category(inventory, id_index, name_index, category_index, stock_index, filter_category):
    filtered_products = []

    for product in inventory:
        categories = product[category_index]
        if filter_category in categories:
            total_stock = sum(product[stock_index])
            filtered_products.append([
                product[id_index],          # Product's ID
                product[name_index],        # Product's name
                categories,                 # Product's categories
                total_stock                 # Total stock available
            ])

    return filtered_products

# Inventory list
inventory = [
    ['P001', 'Smartphone', ['electronics', 'gadgets'], [50, 30, 20]],
    ['P002', 'Laptop', ['electronics', 'computers'], [25, 15, 10]],
    ['P003', 'T-Shirt', ['clothing', 'sportswear'], [100, 200]],
    ['P004', 'Headphones', ['electronics', 'gadgets', 'audio'], [80, 50]],
    ['P005', 'Running Shoes', ['clothing', 'sportswear', 'footwear'], [40, 60]],
    ['P006', 'Blender', ['kitchen', 'home appliances'], [15, 25]],
    ['P007', 'Microwave', ['kitchen', 'home appliances'], [10, 5]],
    ['P008', 'Sofa', ['furniture', 'living room'], [8, 12]],
    ['P009', 'Notebook', ['stationery', 'office'], [150, 200]],
    ['P010', 'Backpack', ['sportswear', 'travel'], [30, 45]]
]

# Call the function with the 'electronics' category
result = get_products_by_category(inventory, 0, 1, 2, 3, 'electronics')

# Print the resulting list
print(result)
```

#### Ejercicio 1: Variables globales

Creemos una variable global, y editemosla desde una función.
- **Hint**: Hagamos una operación de `+=`.

```python
total_stock = 0
```

```python
print(total_stock)
```

```python
def agregar_stock(agregar):
    global total_stock
    local_stock = total_stock

    # vamos a agregar
    local_stock += agregar

    total_stock = local_stock
```

```python
agregar_stock(100)
```

```python
print(total_stock)
```

```python
# Variable global que representa el stock total inicial
total_stock = 500

# Muestra el stock inicial
print(f"Stock inicial: {total_stock}")

# Función para agregar productos al inventario
def add_stock(amount):
    global total_stock  # Accede a la variable global 'total_stock'
    local_stock = total_stock  # Variable local para realizar un seguimiento del stock temporal

    # Agrega la cantidad al stock local
    local_stock += amount
    
    # Actualiza el stock global
    total_stock = local_stock

# Realiza operaciones para agregar stock
add_stock(100)  # Agregar 100 productos
print(f"Stock global actualizado: {total_stock}")

add_stock(250)  # Agregar 250 productos
print(f"Stock global actualizado: {total_stock}")
```

```python
# Inicimos el código aquí
# Inventario inicial de herramientas
inventory = []

# Tarea 1: encapsular la adición de una herramienta
def add_tool(name, category, available=True):
    inventory.append({"name": name, "category": category, "available": available})

# Tarea 2: encapsular la comprobación de disponibilidad
def check_availability(tool_name):
    for tool in inventory:
        if tool["name"] == tool_name:
            print(f"La herramienta '{tool_name}' está disponible: {tool['available']}")
            return
    print(f"La herramienta '{tool_name}' no está en el inventario.")

# Tarea 3: encapsular el préstamo de una herramienta
def lend_tool(tool_name):
    for tool in inventory:
        if tool["name"] == tool_name:
            if tool["available"]:
                tool["available"] = False
                print(f"La herramienta '{tool_name}' ha sido prestada.")
            else:
                print(f"La herramienta '{tool_name}' ya está prestada.")
            return
    print(f"La herramienta '{tool_name}' no está en el inventario.")

# Tarea 4: encapsular el listado de todas las herramientas
def list_tools(status="all"):
    for tool in inventory:
        if status == "all":
            print(f"Nombre: {tool['name']}, Categoría: {tool['category']}, Disponible: {tool['available']}")
        elif status == "available" and tool["available"]:
            print(f"Nombre: {tool['name']}, Categoría: {tool['category']}, Disponible: {tool['available']}")
        elif status == "lent" and not tool["available"]:
            print(f"Nombre: {tool['name']}, Categoría: {tool['category']}, Disponible: {tool['available']}")

# Usar las funciones

# Agregar herramientas al inventario
add_tool("Taladro", "Eléctrica")
add_tool("Martillo", "Manual", available=True)
add_tool("Llave inglesa", "Manual")

# Comprobar la disponibilidad de una herramienta
check_availability("Taladro")
check_availability("Sierra")

# Prestar una herramienta
lend_tool("Taladro")
lend_tool("Llave inglesa")

# Listar todas las herramientas
print("\nTodas las herramientas:")
list_tools("all")

print("\nHerramientas disponibles:")
list_tools("available")

print("\nHerramientas prestadas:")
list_tools("lent")
```
