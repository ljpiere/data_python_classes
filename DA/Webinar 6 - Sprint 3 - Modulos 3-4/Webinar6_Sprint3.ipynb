##### Ejercicio 4.5: Reemplazo de Valores con where()

Cambia la categoría de productos con precio menor a 50 a "Bajo costo".

**Pista:** Utiliza where para reemplazar valores que no cumplan una condición específica.

```python
import pandas as pd
df = pd.read_csv('product_sales.csv')
df.head()
```

```python
df['Categoría'] = df['Categoría'].where(~df['Precio'] < 50, 'Bajo costo')
```
