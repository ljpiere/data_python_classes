```python
import pandas as pd

# Datos de la tabla de plataformas
data = {
    "platform": ["2600", "3DO", "3DS", "DC", "DS", "GB", "GBA", "GC", "GEN", "GG", "N64", "NES", "NG", "PC"],
    "min": [1980, 1994, 2011, 1998, 1985, 1988, 2000, 2001, 1990, 1992, 1996, 1983, 1993, 1985],
    "max": [1989, 1995, 2016, 2008, 2013, 2001, 2007, 2007, 1994, 1992, 2002, 1994, 1996, 2016],
    "active_years": [9, 1, 5, 10, 28, 13, 7, 6, 4, 0, 6, 11, 3, 31],
}

df_platforms = pd.DataFrame(data)
```

```python
df_platforms
```

```python
# Ordenamos por el año de aparición para calcular la diferencia entre lanzamientos
df_platforms = df_platforms.sort_values(by="min")
df_platforms["years_between"] = df_platforms["min"].diff()

# Calculamos los promedios
avg_time_between_new_platforms = df_platforms["years_between"].mean()
avg_lifetime_platforms = df_platforms["active_years"].mean()

print(f"Tiempo promedio entre la aparición de nuevas plataformas: {avg_time_between_new_platforms:.2f} años")
print(f"Tiempo de vida promedio de las plataformas: {avg_lifetime_platforms:.2f} años")
```
